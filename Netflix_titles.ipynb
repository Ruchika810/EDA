##Pie chat

```python
#Create a pie chart of churn column
plt.pie(df1['type'].value_counts(), labels=df1['type'].value_counts().keys())
plt.show()
```

```python
import plotly.express as px
x=df1.groupby(['rating']).size().reset_index(name='counts')
piechart= px.pie(x,names='rating',title='title')
piechart.show()
```

##Bar Plot

```python
df1.columns
```

```python
#Top 10 directors
sns.barplot(x=df1.director.value_counts().head(10), y = df1.director.value_counts().head(10).index,data=df1,orient='h')
plt.show()
```

```python
#Top 10 cast
sns.barplot(x=df1.cast.value_counts().head(10), y = df1.cast.value_counts().head(10).index,data=df1,orient='h')
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.xticks(rotation = 90)
sns.countplot(data = df1, x=df1['director'],palette ='mako',order =df1['director'].value_counts().index[:40])
plt.show()
```
